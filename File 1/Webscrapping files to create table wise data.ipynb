### Loading table:1 data of All the Website

```python
nike1 = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Nike\\nike_Table1.csv")
adidas1 = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Adidas\\adidas_Table1.csv")
puma = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Puma\\puma_Table1.csv")
skechers = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Skechers\\skechers_Table1.csv")
columbia = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Columbia\\colombia_Table1.csv")
woodland = pd.read_csv("E:\\capstone project\\File 1\\Table wise CSV For each website\\Woodland\\Woodland_Table1.csv")
```

```python
# Combining all the dataframe to create one dataframe named table 1
Table_1 = pd.concat([nike1, adidas1, puma, skechers, columbia, woodland], axis=0)
Table_1
```

```python
#Saving table1 as csv
Table_1.to_csv('Table_1.csv', index=False)
```
